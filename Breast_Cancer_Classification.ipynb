##**Splitting the dataset**

```python
# Let's drop the target feature column to create X

X = df_cancer.drop(['target'],axis=1)
```

```python
X
```

```python
#Let's put the target feature in Y

Y = df_cancer['target']
Y
```

```python
#Split the entire data into training and testing dataset

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20,random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

**80% of the data goes into the training dataset**

```python
Y_train.shape
```

```python
Y_test.shape
```

**Remaining 20% of the data goes into the testing dataset**

##**Creating the model and training it**

```python
from sklearn.svm import SVC 
from sklearn.metrics import confusion_matrix, accuracy_score,precision_score,recall_score,f1_score

svc_model = SVC() #Creates an object of the model
svc_model.fit(X_train, Y_train) # Trains the model using X_train and Y_train dataset
```

##**Model evaluation**

```python
Y_predict = svc_model.predict(X_test)
cm = confusion_matrix(Y_test, Y_predict) #Creates a confusion matrix using the original Y_test and the predicted output i.e Y_predict
```

```python
#Plotting a heatmap to represent the confusion matrix

sns.heatmap(cm, annot=True,linewidth=0.5)
```

**Along the X axis we have the predicted classes and along the Y axis the actual classes.**

1.   **True Positive = 66**
2.   **True Negative = 41**
3.   **False Positive = 7**
4.   **False Negative = 0**





```python
act = accuracy_score(Y_test,Y_predict)
print('Test Accuracy is: ',(act*100))
p = precision_score(Y_test,Y_predict)
print('Test Precision is: ',(p*100))
r = recall_score(Y_test,Y_predict)
print('Test Recall is: ',(r*100))
f = f1_score(Y_test,Y_predict)
print('Test F1 Score is: ',(f*100))
```

**Accuracy and precision is good but there is a scope of improvement. We can make changes in our model to improve it's performance.We can observe that the different features have different range of values which might cause this poor performance. Let's try applying normalisation to the data.**

##**Applying Min-Max Normalisation**

min max normalisation.png

```python
# Scaling is applied to each feature to ensure that the values lie with 0-1 range

for column in df_cancer.columns:
    df_cancer[column] = (df_cancer[column] - df_cancer[column].min()) / (df_cancer[column].max() - df_cancer[column].min())    
```

```python
df_cancer
```

```python
sns.scatterplot(x = 'mean radius', y = 'mean smoothness', hue = 'target', data = df_cancer)
```

**If we observe the above scatterplot we can see that both the features are scaled within 0-1 range**

##**Changing the parameters in SVC - Gamma (Kernel Coefficient) and C(Regularization Parameter)**

```python
# Speifying the different values of parameters that we wish to try while training the model

param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel': ['rbf']} 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(),param_grid,refit= True,verbose=3) # The model, param_grid specifying the different parameter values, if refit should be done and verbosity is passed
```

```python
grid.fit(X_train,Y_train) # The model 
```

```python
grid.best_params_ # Returns the best parameters
```

```python
grid.best_estimator_ # Estimator that was chosen by the search, i.e. estimator which gave highest score
```

##**Evaluating the best model**

```python
grid_predictions = grid.predict(X_test) # Call predict on the estimator with the best found parameters.
```

```python
cm = confusion_matrix(Y_test, grid_predictions)
```

```python
sns.heatmap(cm, annot=True,linewidth=0.5)
```

**Along the X axis we have the predicted values and along the Y axis the actual classes.**

```python
act = accuracy_score(Y_test,grid_predictions)
print('Test Accuracy is: ',(act*100))
p = precision_score(Y_test,grid_predictions)
print('Test Precision is: ',(p*100))
r = recall_score(Y_test,grid_predictions)
print('Test Recall is: ',(r*100))
f = f1_score(Y_test,grid_predictions)
print('Test F1 Score is: ',(f*100))
```
